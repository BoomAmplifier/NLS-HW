$$G(s)=\frac{1}{s^3+3s^2+3s+1}$$

**Solution:**    
According to the open-loop system:
$$G(s)=\frac{1}{s^3+3s^2+3s+1}=\frac{1}{(s+1)^3}$$   
$G(s)$ has a 3-order pole which is $s=-1$, the open-loop system is a stable linear system. The open-loop system can be represented in state space form as:
$$\begin{equation}
\left\{
\begin{array}{rCl}
\dot{x_1} &=& x_2 \\
\dot{x_2} &=& x_3 \\
\dot{x_3} &=& -3x_3-3x_2-x_1+u
\end{array}
\right.
\end{equation}
$$
The structure of the system is as follows:

Let $s=j\omega$, we have
$$G(j\omega) = \frac{1}{-3\omega^2+1-j(\omega^3-3\omega)}$$
From $Im[G(j\omega)]=0$, we have the intersection point between $G(j\omega)$ and the real axis is $G(j\sqrt{3})=-0.125+j0$.
Draw the Nyquist Plot of $G(j\omega)$:


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy.interpolate, scipy.optimize
from sympy import *
import control
%matplotlib inline

#Creating a transfer function G
s = control.TransferFunction.s

K0 = 1.0
# T = 10
G = K0 / (s**3+3.0*(s**2)+3.0*s+1.0)

control.nyquist(G)

plt.grid(True)
plt.title('Nyquist Diagram of G(s)')
plt.text(-0.125,0.1*1.01,(-0.125,0))
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.ylim(-1,1)
plt.xlim(-0.3,1.1)
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
x1 = np.linspace(-10,-0,5,endpoint=True)
y1 = 0*x1
ax.plot(x1,y1)
control.nyquist(G)
plt.grid(True)
plt.title('Nyquist Diagram of G(s)')
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.ylim(-.1,.2)
plt.xlim(-0.25,.1)
plt.show()
```
